```python
import cv2 as cv
import matplotlib.pyplot as plt 
import matplotlib.image as img
```

```python
image1 = cv.imread("Car.jpg")
image2 = cv.imread("Scenery.jpg")
```

```python
#resizing both the images to same resolution

down_width = 300
down_height = 400
down_points = (down_width, down_height)
image1 = cv.resize(image1, down_points, interpolation= cv.INTER_LINEAR)
image2 = cv.resize(image2, down_points, interpolation= cv.INTER_LINEAR)
```

```python
fig = plt.figure(figsize=(10, 7)) 
rows = 1
columns = 2

#Adding subplot to the image
fig.add_subplot(rows, columns, 1) 
# showing image 
plt.imshow(image1) 
plt.axis('off') 
plt.title("Image 1")


#Adding subplot to the image
fig.add_subplot(rows, columns, 2) 
# showing image 
plt.imshow(image2) 
plt.axis('off') 
plt.title("Image 2")
```

```python
combinedImage = cv.hconcat([image1, image2]) 
```

```python
plt.imshow(combinedImage)
plt.axis("off")
plt.title("Combined Image")
plt.savefig("combined.png")
```
