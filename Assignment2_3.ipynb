```python
import cv2 as cv
import matplotlib.pyplot as plt 
import matplotlib.image as img
```

```python
testImage = cv.imread("scenery.jpg")
```

```python
plt.imshow(testImage)
```

```python
#selecting region to crop
x_start = 800
x_end = 2500
y_start = 1500
y_end = 3700
```

```python
croppedImage = testImage[y_start:y_end, x_start:x_end]
```

```python
plt.imshow(croppedImage)
```

```python
fig = plt.figure(figsize=(10, 7)) 
rows = 1
columns = 2

#Adding subplot to the image
fig.add_subplot(rows, columns, 1) 
# showing image 
plt.imshow(testImage) 
plt.axis('off') 
plt.title("Orginal Image")


#Adding subplot to the image
fig.add_subplot(rows, columns, 2) 
# showing image 
plt.imshow(croppedImage) 
plt.axis('off') 
plt.title("Extracted Image")
```
